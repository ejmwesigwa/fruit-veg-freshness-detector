```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from collections import Counter
import random
import pandas as pd


# Define path
data_path = "data/raw/Fruit And Vegetable Diseases Dataset/"

# List all class folders
class_dirs = sorted([d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))])
print(f'Total Classes: {len(class_dirs)}')
print(f'Classes: {class_dirs}')

class_counts = {}
for class_dir in class_dirs:
    class_path = os.path.join(data_path, class_dir)
    count = len(os.listdir(class_path))
    class_counts[class_dir] = count

# Display as DataFrame
df_counts = pd.DataFrame(list(class_counts.items()), columns=['Class', 'Image_Count'])
df_counts = df_counts.sort_values(by='Image_Count', ascending=False).reset_index(drop=True)
print(df_counts)

# Barplot
plt.figure(figsize=(14, 6))
sns.barplot(x='Class', y='Image_Count', data=df_counts)
plt.xticks(rotation=90)
plt.title('Image Count per Class')
plt.tight_layout()
plt.show()

```

```python
dims = []
for class_dir in class_dirs:
    class_path = os.path.join(data_path, class_dir)
    for img_file in os.listdir(class_path):
        img_path = os.path.join(class_path, img_file)
        try:
            img = Image.open(img_path)
            dims.append(img.size)  # (width, height)
        except:
            pass

# Convert to DataFrame and describe
df_dims = pd.DataFrame(dims, columns=['Width', 'Height'])
print(df_dims.describe())

# Plot distribution
plt.figure(figsize=(12, 5))
sns.histplot(df_dims['Width'], bins=30, kde=True, color='skyblue', label='Width')
sns.histplot(df_dims['Height'], bins=30, kde=True, color='salmon', label='Height')
plt.title('Image Dimension Distribution')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(3, 4, figsize=(12, 9))
for ax in axes.flatten():
    class_dir = random.choice(class_dirs)
    class_path = os.path.join(data_path, class_dir)
    img_file = random.choice(os.listdir(class_path))
    img_path = os.path.join(class_path, img_file)
    img = Image.open(img_path)
    ax.imshow(img)
    ax.set_title(class_dir)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
